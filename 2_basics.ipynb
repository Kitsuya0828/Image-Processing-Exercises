```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

img = np.zeros([1, 3, 3], dtype=np.uint8)

img[0, 0, 0] = 255  # R
img[0, 1, 2] = 255  # B
img[0, 2, 1] = 255  # G

plt.imshow(img) # 
io.imsave('results/RBG.png', img) # 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

img = np.zeros([100, 300, 3], dtype=np.uint8)

img[:, :100, 0] = 255  # R
img[:, 100:200, 2] = 255  # B
img[:, 200:, 1] = 255  # G

plt.imshow(img) # 
io.imsave('results/RBG_large.png', img) # 
```
